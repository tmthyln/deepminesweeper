# NxM Simulations

```python
import matplotlib.pyplot as plt

from project.runner import AutoGameRunner
from game.mineseeders import RandomSeeder

T = 200
shape = (100, 100)
```

## Simulation

```python
ticks, oos, perco, msd = [], [], [], []

for trial in range(1, T + 1):
    runner = AutoGameRunner(shape, seeder=RandomSeeder(shape))
    runner.run()
    
    ticks.append(runner.ticks)
    oos.append(runner.oos)
    perco.append(runner.perco)
    msd.append(runner.msd)
```

## Statistics

```python
plt.plot(ticks)
plt.show()
plt.plot(oos)
plt.show()
plt.plot(perco)
plt.show()
plt.plot(msd)
plt.show()
```
